### We are gonna scrape https://github.com/topics

- I will scrape the popular topics available on github


### Using requests library to fetch the page

```python
import requests 
```

```python
topic_url = requests.get('https://github.com/topics')
topic_url.status_code
```

```python
topic_url = requests.get('https://github.com/topics').text
```

### Using BeautifulSoup to extract information

```python
from bs4 import BeautifulSoup


soup = BeautifulSoup(topic_url,'html.parser')

print(soup)
```

```python
title_text = []

topic_title = soup.find_all('p', class_='f3 lh-condensed mb-0 mt-1 Link--primary')

for tag in topic_title:
    title_text.append(tag.text)
    
title_text
```

```python
title_des = []

topic_desc = soup.find_all('p',class_='f5 color-fg-muted mb-0 mt-1')

for desc in topic_desc:
    title_des.append(desc.text.strip())
    
title_des
```

```python
topic_title_url = soup.find_all('a', class_='d-flex no-underline')
topic_title_url[0]['href']
```

```python
topics_url = []
base_url = 'https://github.com'

for url in topic_title_url:
    topics_url.append(base_url + url['href'])
    
topics_url
```

### Saving data into csv

```python
import pandas as pd


topic_dict = {'Title': title_text,'Description':title_des,'URL':topics_url}

topics_df = pd.DataFrame(topic_dict)

topics_df
```
